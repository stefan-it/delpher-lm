```python
import numpy as np
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
with open('./stats/delpher_year_stats.pickle', 'rb') as f_p:
    delpher_char_stats = pickle.load(f_p)
```

```python
df_delpher_stats = pd.DataFrame(delpher_char_stats.items(), columns=['year', 'length'])
df_delpher_stats["year"] = pd.to_numeric(df_delpher_stats["year"])

plt.figure(figsize=(50, 6))
delpher_plot = sns.barplot(x="year", y="length", data=df_delpher_stats.sort_values(by=['year']))

delpher_plot.set_title("Number of chars per year (Delpher Corpus)")
delpher_plot.set(ylabel="Number of chars")
delpher_plot.set(xlabel="Year")
delpher_plot.set_xticklabels(delpher_plot.get_xticklabels(), rotation=90)

fig = delpher_plot.get_figure()
fig.savefig("./figures/delpher_corpus_stats.png")
```
